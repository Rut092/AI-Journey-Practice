# Decision Trees

```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
```

```python
print(iris)
```

```python
X = iris.data[:,[2,3]]
y = iris.target
print('Class Labels : ',np.unique(y))
```

```python
print(iris.target_names)
```

When you set stratify=y, where y is your target variable, the function splits the data so that both the training and test sets maintain the same class distribution as the full dataset.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 1, test_size = 0.3,stratify = y)
```

```python
'''
NumPy's bincount function counts the number of occurrences of each value in a 1-dimensional array of non-negative integers.
np.bincount([0, 1, 2, 3, 4, 4, 6])
>>>   array([1, 1, 1, 1, 2, 0, 1])
'''
print('Labels counts in y:',np.bincount(y))
print('Labels counts in y_train:',np.bincount(y_train))
print('Labels counts in y_test:',np.bincount(y_test))
```

```python
'''
Instead of opening plots in a separate window or displaying a text representation of the plot object,
 %matplotlib inline embeds the plot as a static image directly in the notebook.
 '''
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions

tree = DecisionTreeClassifier(criterion = 'entropy', max_depth=2,random_state=1)
tree.fit(X_train,y_train)

plot_decision_regions(X_train, y_train,tree)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.title('Decision Tree')
plt.show()
```
