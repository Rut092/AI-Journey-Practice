ðŸš  The Core Concept: The Foggy Mountain
Imagine you are standing at the top of a mountain in thick fog. You want to reach the village at the very bottom (the Global Minimum of error). You canâ€™t see the village, but you can feel the slope under your feet.

Feel the slope: This is the Gradient (the derivative).

Take a step: This is the Update. The size of your step is the Learning Rate.

Repeat: You keep doing this until the ground feels flat. At this point, you have "converged."

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(w):
    return w**2
```

```python
def gradient_function(w):
    return 2*w   # on differentiation we get w^2 -> 2*w
```

```python
print(np.linspace(-10,10,100)) # generate evenly space array from -10 to 10
```

```python
w = 10
learning_rate = 0.01
epoch = 20
history = []


for i in range(epoch):
    w = w - learning_rate * gradient_function(w)
    history.append(cost_function(w))


weights = np.linspace(-11,11,100)
plt.plot(weights,cost_function(weights),color='black', label='The Valley (Error)')
plt.scatter(history,[cost_function(w) for w in history],color='red', label='Local Minimum')
plt.title(f"Final Weight: {w:.4f}")
plt.legend()
plt.show()
```
