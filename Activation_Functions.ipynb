## Forward Pass (Core Function)

```python
import numpy as np

z = np.array([-3.0, -1.0, 0.0, 1.5, 4.0])

def sigmoid(x):
    return 1/(1+ np.exp(-x))

def ReLU(x):
    return np.maximum(0,x)

def SoftMax(x):
    # We subtract np.max(x) for numerical stability so e^x doesn't explode to infinity
    exp_x = np.exp(x - np.max(x))
    return exp_x/np.sum(exp_x)

print(f"Sigmoid: {sigmoid(z)}")
print(f"ReLU: {ReLU(z)}")
print(f"SoftMax: {SoftMax(z)}")
```

## Backward Pass Functions

```python
def relu_derivative(x):
    return np.where(x>0,1.0,0.0)

def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))

print(f"ReLU Derivative: {relu_derivative(z)}")
print(f"Sigmoid Derivative: {sigmoid_derivative(z)}")
```
