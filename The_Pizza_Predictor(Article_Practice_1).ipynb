# Pizza Price Predictor based on Toppings

```python
# Preparation of data
X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([7, 9, 11, 13, 15])
# Hidden Rule 2*x + 5

model = LinearRegression()
model.fit(X,Y)
```

```python
prediction = model.predict([[7]])
print(f"A Pizza with 7 toppings should cost around {prediction[0]:.2f} Rupees")
```

# Practicing on more Data and Plotting the Graph

```python
# Create 100 random points for 'x' 100 rows and 1 column
# The np.random.rand() function in NumPy generates random values from a uniform distribution over the interval [0, 1)
X = np.random.rand(100,1)

# Create 'y' based on the formula, but add some "noise" (randomness)
#np.random.randn generates random numbers from a standard normal (Gaussian) distribution with a mean of 0 and a variance of 1,
#which means values are more likely to be close to zero, with decreasing probability as values move further from the mean.
Y = 4 + 3*X + np.random.randn(100,1)
```

```python
model.fit(X,Y)
print(f"Model's Coefficient: {model.coef_}")
print(f"Model's Intercept: {model.intercept_}")
```

```python
# Let's look at the data
plt.scatter(X,Y,color = 'red')
plt.plot(X, model.predict(X))
plt.title("Testing of Linear Model Ax+ B")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
